```python
def get_sales_count(symbol, letter_no, year):
    try:
        financial_year = _get_financial_year(symbol, year)
        if financial_year == "No Data found":
            raise ValueError("Financial year not found for the given symbol.")
        response = get_reports(symbol, financial_year)
        date = digits.fa_to_en(re.findall(r'\d{4}/\d{1,2}/\d{1,2}' ,response['Letters'][letter_no]['Title'])[0])
        urlp = urlparse(response['Letters'][letter_no]['Url'])
        params = parse_qs(urlp.query)
        params['SheetId'] = 20
        url = f"{BASE_URL}{urlp.path}"
        resp = requests.get(url, params=params)
        statement = resp.text
        pattern = r"var datasource = (.*?);"
        match = re.search(pattern, statement)
        if match:
            text = match.group(1)
        records = []
        records.append(
            (statement, text))
        for letter, data in records:
            continue
            #print(data)
        items = json.loads(data)['sheets']
        cells = items[0]['tables'][1]['cells']
        cells_ = [(i['columnSequence'], i['rowSequence'], i['value'], i['periodEndToDate']) for i in cells]
        filtered_cells = [(i[0], i[1], i[2]) for i in cells_ if i[-1] == '' or i[-1] == (date)]
        df = pl.from_records(filtered_cells, schema=["col", "row", "value"], orient="row")
        df = df.pivot(values="value", on="col", index="row").sort("row")

        index = [i for i, v in enumerate(df.filter(pl.col("1") == "نوع کالا").row(0)) if characters.ar_to_fa(str(v)) == "نوع کالا"]
        matching_columns = [col for col in df.columns if "تعداد فروش" in df[col].to_list()]
        selected = df.select([str(index[0]), str(matching_columns[0])]).drop_nulls()

        selected = selected.with_columns(
            pl.col(str(index[0]))
            .map_elements(characters.ar_to_fa, return_dtype=pl.Utf8)
            .alias("item"),

            pl.col(str(matching_columns[0]))
            .replace("", None)
            .cast(pl.Int64)
            .alias("sales_count")
        ).drop([str(index[0]), str(matching_columns[0])])
        selected = selected.filter(
            (pl.col("item") != "") & (pl.col("item").is_not_null())
        )
        return selected
    except Exception as e:
        print(f"Error processing symbol {symbol} for letter {letter_no} in year {year}: {str(e)}")
        return None
```

```python
#HW1
BASE_URL = "https://www.codal.ir"
SEARCH_URL = "https://www.codal.ir/ReportList.aspx?search"
API_URL = "https://search.codal.ir/api/search/v1"

def normalize_fs_item(w: str) -> str:
   dict_ = {
      " ": "",
      "–": "",
      "(": "",
      ")": "",
      "،": "",
      "ي": "ی",
      "ى": "ی",
      "ك": "ک",
      "آ": "ا",
      "\u200f": "",
      "\u200c": ""
   }
   return w.translate(str.maketrans(dict_))

def _create_dataframe(url: str, params: dict) -> pl.DataFrame:
    resp = requests.get(url, params=params)
    statement = resp.text
    pattern = r"var datasource = (.*?);"
    match = re.search(pattern, statement)
    records = []
    validation_error = []
    if match:
        text = match.group(1)
        try:
            records.append(
                (statement, text)
            )
        except Exception as e:
            validation_error.append((statement, str(e)))
            return pd.DataFrame()

        items = json.loads(records[0][1])['sheets']
        cells = items[0]['tables'][0]['cells']
        cells_ = [(i['columnSequence'], i['rowSequence'], i['value']) for i in cells]
        df = pl.from_records(cells_, schema=["col", "row", "value"], orient="row")
        df = df.pivot(values="value", on="col", index="row")
        df_filter = df.filter(pl.col("1") == "شرح")
        index = []
        for ind, value in enumerate(df_filter.row(0)):
            if normalize_fs_item(str(value)) == 'شرح':
                index.append(ind)
        selected = df.select([str(index[0]), str(index[0]+1)])
        selected = selected.rename(selected.head(1).to_dicts().pop()).drop_nulls()
        selected = selected.filter(~pl.col(selected.columns[1]).is_in([""]))
        selected = selected.filter(pl.col(selected.columns[0]) != "شرح")
        selected = selected.filter(pl.col(selected.columns[0]) != "1000")
        return selected


def create_report_table(symbol:str, length:int, letter_type: int, letter_no: int, sheet_id: int = 0) -> pl.DataFrame:
    """Get search result based on length and letter type of a symbol and create dataframe of nth letter number.

    Args:
        symbol (str): symbol of a company
        length (int): length of a report, -1 means all lengths and rest is from 1 to 12
        letter_type (int): 'اطلاعات و صورتهای مالی میاندوره ای': 6 , 'صورت وضعیت پورتفوی': 8 ,'گزارش فعالیت ماهانه': 58 
        letter_no (int): nth report returns
        sheet_id (int): 0 for BalanceSheet and 1 for IncomeStatement if letter_type is 6

    Returns:
        pd.DataFrame: created dataframe of a report
    """
    
    try:
        page_no = int(letter_no/20) + 1
        url = f"https://search.codal.ir/api/search/v2/q?&Audited=true&AuditorRef=-1&Category=-1&Childs=false&CompanyState=-1&CompanyType=-1&Consolidatable=true&IsNotAudited=false&Length={length}&LetterType={letter_type}&Mains=true&NotAudited=true&NotConsolidatable=true&PageNumber={page_no}&Publisher=false&ReportingType=-1&TracingNo=-1&search=false&Symbol={symbol}"
        payload = {}
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:138.0) Gecko/20100101 Firefox/138.0',
            'Accept': 'application/json, text/plain, */*',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate, br, zstd',
            'Origin': 'https://www.codal.ir',
            'Connection': 'keep-alive',
            'Referer': 'https://www.codal.ir/',
            'Sec-Fetch-Dest': 'empty',
            'Sec-Fetch-Mode': 'cors',
            'Sec-Fetch-Site': 'same-site',
            'DNT': '1',
            'Sec-GPC': '1',
            'Cookie': 'TS018fb0f7=01f9930bd24c8c30357f04a2d3b99189ea0e255496515c3af6b226a4e1fd75ccfcb8b9eddde1066c4174f70391afd4edca265b71d3; Unknown=1076170924.20480.0000'
        }

        response = requests.request("GET", url, headers=headers, data=payload)
        response = response.json()
        if len(response['Letters']) == 0:
            print("Report is not available")
            url = f"https://search.codal.ir/api/search/v2/q?&Audited=true&AuditorRef=-1&Category=-1&Childs=false&CompanyState=-1&CompanyType=-1&Consolidatable=true&IsNotAudited=false&Length={length}&LetterType={letter_type}&Mains=true&NotAudited=true&NotConsolidatable=true&PageNumber=1&Publisher=false&ReportingType=-1&TracingNo=-1&search=false&Symbol={symbol}"
            response = requests.request("GET", url, headers=headers, data=payload)
            response = response.json()
            print("Max number of reports:", response['Total'])
            return pd.DataFrame()
        letter_no = abs(20-letter_no) if letter_no > 19 else letter_no
        urlp = urlparse(response['Letters'][letter_no]['Url'])
        params = parse_qs(urlp.query)
        params['SheetId'] = sheet_id
        url = f"{BASE_URL}{urlp.path}"
        return _create_dataframe(url, params)
    except Exception as e:
        print((e))

def get_category_letter_types():
  url = "https://search.codal.ir/api/search/v1/categories"

  payload = {}
  headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:138.0) Gecko/20100101 Firefox/138.0',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br, zstd',
    'Origin': 'https://www.codal.ir',
    'Connection': 'keep-alive',
    'Referer': 'https://www.codal.ir/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'Cookie': 'TS018fb0f7=01f9930bd234cd7faf40adcddf524ae822895234dc1cdcf2e81d21f0cd71caa62ca24e791bf43de8e7ee40c45c9afec0cf38d2ca8c; Unknown=1076170924.20480.0000'
  }

  response = requests.request("GET", url, headers=headers, data=payload)

  response = response.json()
  category_dict = {}
  letter_type_dict = {}
  for items in response[1:]:
      category_dict[items['Name']] = items['Code']
      for letters in items['PublisherTypes']:
          if letters['Name'] == 'ناشران':
              for letter_types in letters['LetterTypes']:
                  letter_type_dict[letter_types['Name']] = letter_types['Id']
  return category_dict, letter_type_dict

create_report_table("ارفع", 12, 6, 0, 0)
```
